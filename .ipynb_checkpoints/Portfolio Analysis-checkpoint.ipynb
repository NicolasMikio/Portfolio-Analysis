# Análise de Portfólio

Construção de Portfólio via Maximum Sharpe Ratio Portfolio, Global Minimum Volatility Portfolio, Equaly Equally Weighted Portfolio

```python
#Importa as bibliotecas
import yfinance as yf
import math
import matplotlib.pyplot as plt
import numpy as np
from time import time
import pandas as pd
from datetime import date
import edhec_risk_kit as erk
plt.style.use('dark_background')
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

O código e parte das funções foram desenvolvidas por mim, mas a maoria das funções foram extraídas do curso Introduction to Portfolio Construction and Analysis with Python da EDHEC Business School ministrada no Coursera.

```python
#Importa os dados do Yahoo Finance
data = yf.download("ITSA4.SA PETR4.SA BRML3.SA PSSA3.SA", start="2018-01-01", interval = "1d")
display(data)
```

```python
#Calcula os retornos das série
x = erk.calcula_retornos(data)
x
```

```python
x.std()
```

```python
#Calcula os retornos anualizados
xano = erk.annualize_rets(x, 252)
xano
```

```python
#Calcula a matriz de covariância anualizada
cov = x.cov()*252
cov
```

```python
#Calcula a Taxa Livre de Risco
cdi = erk.consulta_bc(12)
cdi_recente = cdi.iloc[-1]
r = (1 + cdi_recente/100).prod()**(252)-1
r
```

## Minimizando a volatilidade para dado retorno

```python
# Pesos para um portfolio com 7% de retorno esperado, minimizando a volatilidade
array2 = erk.minimize_vol(0.070, xano, cov).round(2)
df = pd.DataFrame(array2)
df.index = xano.index
df = df.rename(columns={0: "Pesos"})
df
```

```python
print("Portfolio return:", erk.portfolio_return(array2, xano).round(2))
print("Portfolio volatility:", erk.portfolio_vol(array2, cov).round(2))
```

## Global Minimum Variance Portfolio

```python
gmv2 = erk.gmv(cov)
pesos = pd.DataFrame(gmv2)
pesos.index = xano.index
pesos = pesos.rename(columns={0: "Pesos"})
pesos
```

```python
print("Global Minimum Variance return:", erk.portfolio_return(gmv2, xano).round(2))
print("Global Minimum Variance volatility:", erk.portfolio_vol(gmv2, cov).round(2))
```

## Maximum Sharpe Portfolio

```python
# Pesos do Maximum Sharpe Portfolio
array = erk.msr(r, xano, cov).round(2)
msrpesos = pd.DataFrame(array)
msrpesos.index = xano.index
msrpesos = msrpesos.rename(columns={0: "Pesos"})
msrpesos
```

```python
#Plota a fronteira de Markowitz e o Capital Market Line
erk.plot_ef(40, xano, cov, show_cml=True, riskfree_rate = r)
```

```python
print("Maximum Sharpe Portfolio return:", erk.portfolio_return(array, xano).round(2))
print("Maximum Sharpe Portfolio volatility:", erk.portfolio_vol(array, cov).round(2))
```

## Equally Weighted Portfolio

```python
w = np.array([.25, 0.25, 0.25, 0.25])

print("Equally Weighted Portfolio return:", erk.portfolio_return(w, xano).round(2))
print("Equally Weighted Portfolio volatility:", erk.portfolio_vol(w, cov).round(2))
```
